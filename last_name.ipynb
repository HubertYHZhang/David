```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ifreesite.com/scriptures/hundred-family-surnames-01.htm"
page = requests.get(url)
soup = BeautifulSoup(page.content, 'html.parser')

io_content = soup.find('div', {'id': 'io_content'})
tables = io_content.find_all('table')

last_name_pinyin = []
last_name_canton = []

for table in tables:
    rows = table.find_all('tr')
    for row in rows:
        cells = row.find_all('td')
        if len(cells) > 4:  # Check if row has at least 5 cells
            last_name_canton.append(cells[4].get_text())
            last_name_pinyin.append(cells[2].get_text())

i = 0
while i <= 810:
    last_name_canton[i]='DELETE'
    last_name_pinyin[i]='DELETE'
    i += 81
for i in [893,974,1023]:
    last_name_canton[i]='DELETE'
    last_name_pinyin[i]='DELETE'
while 'DELETE' in last_name_canton or 'DELETE' in last_name_pinyin or '' in last_name_canton:
    if 'DELETE' in last_name_canton:
        last_name_canton.remove('DELETE') 
    if 'DELETE' in last_name_pinyin:
        last_name_pinyin.remove('DELETE') 
    if '' in last_name_canton:
        last_name_canton.remove('') 

for li in [last_name_canton,last_name_pinyin]:
    for i,cell in enumerate(li):
        li[i] = li[i].upper()
        
last_name_canton_new = []
for cell in last_name_canton:
    for subcell in cell.split('/'):
        last_name_canton_new.append(subcell)
        
last_name_canton = list(set(last_name_canton))
last_name_pinyin = list(set(last_name_pinyin))

with open('last_name_canton.txt','w',encoding='utf-8') as file:
    for item in last_name_canton:
        file.write(str(item)+'\n')
        
with open('last_name_pinyin.txt','w',encoding='utf-8') as file:
    for item in last_name_pinyin:
        file.write(str(item)+'\n')
```
